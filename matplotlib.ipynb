# chapter 2 



markers are like the coordinates of the points 

types of markers 
1. basic dot one 

```python
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.plot(x,y,marker="." )
plt.show()
```

2. marker with circle


```python
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.plot(x,y,marker="o" )
plt.show()
```

3. arrow type of marker
 

```python
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.plot(x,y,marker="v" )
plt.show()
```

4. star marker graph 

```python
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.plot(x,y,marker="*" )
plt.show()
```

we can also control the size of the marker using the markersize function 

```python
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.plot(x,y,marker="o",markersize=20)
plt.show()
```

we can also write marker size as ms for short 

```python
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.plot(x,y,marker=".",ms=20)
plt.show()
```

changing the color of the marker

```python
# changing the color of the marker 
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan")
plt.show()
```

a shortform to change the color mfc which means markerfacecolor can also be used 

```python
# changing the color of the marker face
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan")
plt.show()
```

to change the color of the marker`s edge use markereedge color or mec for short

```python
# changing the color of the marker 
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan",markeredgecolor="red")
plt.show()
```

customizing the lines 

1. line style 


5 types of line styles are available 
1. normal solid 
2. dashed -- dashed line genrated
3. dotted -- dotted line generated
4. dashdot -- dah and dot ka combi line is generated
5. none which just gives points

```python
# changing the color of the marker 
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan",linestyle="dashed")
plt.show()
```

```python
# changing the color of the marker 
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan",linestyle="dotted")
plt.show()
```

```python
# changing the color of the marker 
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan",linestyle="dashdot")
plt.show()
```

```python
# changing the color of the marker 
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan",linestyle="None")
plt.show()
```

line width customization 

can be done using the linewidth keyword 
The usual value is 1 

```python
# changing the color of the marker 
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan",linestyle="solid",linewidth=4)
plt.show()
```

```python
# changing the color of the marker 
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan",linestyle="dashed",color="pink")
plt.show()
```
