```python
x = np.array([2023,2024,2025,2026])
y1 = np.array([12,89,65,74])
y2 = np.array([45,46,54,40])
y3 = np.array([10,11,12,13])
# creating a dict from the prev customization values 
line_style = dict(marker=".",
                  markersize=20,
                  markerfacecolor="cyan",
                  linestyle="solid",
                  linewidth=4)
plt.plot(x,y1,color="red",**line_style)
plt.plot(x,y2,color="green",**line_style)
plt.plot(x,y3,color="black",**line_style)
```

# chapter 3 Lables and title in matplotlib  

customizing the title in plots

1. font size : to set the font size
 

```python
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.title("Class sizes",fontsize=20)
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan")
plt.show()
```

2. family : this changes the font family of the font 

```python
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.title("Class sizes",
          fontsize=20,
          family ="arial")
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan")
plt.show()
```

3. font weight = "bold" : used to make font bolder for the title

```python
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.title("Class sizes",
          fontsize=20,
          family ="arial",
          fontweight ="bold")
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan")
plt.show()
```

4. color: used to color the title

```python
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.title("Class sizes",
          fontsize=20,
          family ="arial",
          color="red")
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan")
plt.show()
```

lables  :- 


these are the quantities which we represent at the coordinates eg time space etc 

there are two lables normally 


1. xlabel ==> used to set the label in x axis 
2. ylabel ==> used to sset the label in y axis 

can use all the customizations which are used in makking titles


```python
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.title("Class sizes",
          fontsize=20,
          family ="arial",
          color="red")
plt.xlabel("year",
        fontsize=20,
          family ="arial",
          color="blue")
plt.ylabel("subsribes",
        fontsize=20,
        family ="arial",
        color="green")
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan")
plt.show()
```

ticks and its customization

1. usually they show with a .5 increment if you dont want it then do like this 

```python
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.title("Class sizes",
          fontsize=20,
          family ="arial",
          color="red")
plt.xlabel("year",
        fontsize=20,
          family ="arial",
          color="blue")
plt.ylabel("subsribes",
        fontsize=20,
        family ="arial",
        color="green")
plt.xticks(x) # this will only modifiy the ticks on the x axis as it says tself as xticks
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan")
plt.show()
```

```python
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.title("Class sizes",
          fontsize=20,
          family ="arial",
          color="red")
plt.xlabel("year",
        fontsize=20,
          family ="arial",
          color="blue")
plt.ylabel("subsribes",
        fontsize=20,
        family ="arial",
        color="green")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan")
plt.show()
```

if we want to customize our ticks then callthe ticks_param function

```python
x = [2023,2024,2025,2026]
y = [38,40,57,68]
plt.title("Class sizes",
          fontsize=20,
          family ="arial",
          color="red")
plt.xlabel("year",
        fontsize=20,
          family ="arial",
          color="blue")
plt.ylabel("subsribes",
        fontsize=20,
        family ="arial",
        color="green")
plt.plot(x,y,marker=".",markersize=20,markerfacecolor="cyan")
plt.xticks(x)
plt.yticks(y)
plt.tick_params(axis="both",colors="darkblue")
plt.show()
```
